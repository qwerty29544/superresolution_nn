```python
class DetectionSystem:
    def __init__(self, K, d, k):
        self._K = K
        self._k = k
        self._d = d
        self._M = 2*K + 1
        self.ND = 100 / (self.M)
        
    def F(self, t, alpha = 0):
        result = 0
        for n in np.arange(1, self._K + 1):
            result += np.cos(self._k * self._d * n * (t - alpha))
        return result * 2 * self.ND + self.ND
```

```python
class Signal_I:
    def __init__(self, distance_between = 0.1, maximum = 100, variance = 0.05):
        self._max = maximum
        self._var = variance
        self._between = distance_between
            
    def func_gauss(self, alpha):
        return 1/(2 * np.pi * np.sqrt(self._var)) * \
                (np.exp((alpha - self._between)**2/(2 * self._var)) + 
                 np.exp((alpha + self._between)**2/(2 * self._var)))
```

```python
alpha_grid = np.linspace(-1, 1, 101)
```

```python
signal = Signal_I() 
plt.plot(alpha_grid, signal.func_gauss(alpha_grid))
```
